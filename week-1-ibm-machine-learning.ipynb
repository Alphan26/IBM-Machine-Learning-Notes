<h1> What is Artificial Intelligence </h1>

<p>Artificial Intelligence is a branch of computer science dealing with the simulation of intelligent behavior in computers. Machines mimic cognitive functions such as learning and problem solving.  </p>

<h2> Machine Learning </h2>

<p> It learns the patterns of the model by exposing lots of data. </p>

<p> Supervised Learning : Has a target column and it makes predictions. Fraud detection could be an example. </p>

<p> Unsupervised Learning : Does not have a target column. It tries to find structure in the data. Customer segmentation can be an example of that. You can find groupings inside your dataset. </p>

<p> If we think about the images, each pixel can be considered as a feature. But if we think about each pixel there will be large data. Also we can't make relationships with those pixels between each other. That's why we need something else. This is where Deep Learning comes into play. Deep learning can learn these relationships on its own and more suitable for image classification </p>

<p> Deep learning is the cutting edge (latest technology) in larger datasets.  </p>

<p> If you don't have a stable dataset i.e your data is changing much, ML algorithms will perform better. </p>
